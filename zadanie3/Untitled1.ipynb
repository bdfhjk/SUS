```python
import os
import random
import pandas as pd

df = pd.DataFrame(columns=('name', 'a', 'b', 'c', 'd', 'e'))
iter = 0
l = os.listdir("literki")
for item in l:
    path = "literki/" + item
    #print path
    l2 = os.listdir(path)
    a = random.randrange(0, 101, 1)
    b = random.randrange(0, 101, 1)
    c = random.randrange(0, 101, 1)
    d = random.randrange(0, 101, 1)
    e = random.randrange(0, 101, 1)
    for item2 in l2:
        fa = 1.0 * a / 100
        fb = 1.0 * b / 100
        fc = 1.0 * c / 100
        fd = 1.0 * d / 100
        fe = 1.0 * e / 100
 
        da = random.randrange(0,10000,1)
        db = random.randrange(0,10000,1)
        dc = random.randrange(0,10000,1)
        dd = random.randrange(0,10000,1)
        de = random.randrange(0,10000,1)
        
        fa += 1.0 * da / 2000000
        fb += 1.0 * db / 2000000
        fc += 1.0 * dc / 2000000
        fd += 1.0 * dd / 2000000
        fe += 1.0 * de / 2000000

        df.loc[iter] = [item2, fa, fb, fc, fd, fe]
        iter = iter + 1
```

```python
print 'starting clustering'
from sklearn.cluster import AgglomerativeClustering
cl = AgglomerativeClustering(n_clusters=59)
df2 = cl.fit_predict(df.drop('name', axis=1))
print 'Agregated, starting saving'

s = pd.Series(df2, index=range(1,len(df2)+1))
df3 = pd.concat([df, s], axis=1, ignore_index=True)

path = "category"
os.mkdir(path)
for i in range(0,60):
    path = "category/%d" % i
    os.mkdir(path, 0755 );
```

```python
from shutil import copyfile
for r in range(1,6169):
    image_name = df3.loc[r][0]
    image_cat = df3.loc[r][df3.shape[1] - 1]
    image_path_in = 'input/%s' % image_name
    image_path_out = 'category/%d/%s' % (image_cat, image_name)
    copyfile(image_path_in, image_path_out)
    if r % 500 == 0:
        print image_path_in + " " + image_path_out
print 'Saved.'

print 'Prepare submission'
df.drop('name', axis=1).to_csv('submission.csv', index=False, header=False)
print 'Submission saved.'
```
