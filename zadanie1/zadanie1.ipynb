Pracę nad zadaniem rozpocząłem od wczytania danych do postaci DataFrame, wyświetlenia zbioru danych w celu weryfikacji jego poprawności oraz podziału na zbiór treningowy i testowy.

Następnie dodałem dodatkowe zmienne będące produktem iloczynów kartezjańskim wejściowych zmiennych warunkowych w celu znalezienia zależności nieliniowych między danymi.

```python
import sys
import csv
import numpy as np
import pandas as pd
from itertools import product
from matplotlib import pyplot as plt
from sklearn import linear_model, grid_search, tree
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier

reload(sys)
sys.setdefaultencoding("utf-8")
%matplotlib inline
```

```python
df = pd.read_csv('dane.csv')

np.random.randn(42)
msk = np.random.rand(len(df)) < 0.8

train_df = df[msk]
test_df  = df[~msk]

original_features = range(1,11)

train_X = train_df[['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10']]
test_X = test_df[['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10']]

train_Y = train_df['D']
test_Y = test_df['D']

for (f1, f2) in product(original_features, original_features):
    train_X['C%s_%s' % (f1, f2)] = train_X['C%s' % f1] * train_X['C%s' % f2]
    
for (f1, f2) in product(original_features, original_features):
    test_X['C%s_%s' % (f1, f2)] = test_X['C%s' % f1] * test_X['C%s' % f2]
```

Sprawdzam, czy istnieje korelacja między klasą decyzyjną a pozostałymi kolumnami zbioru danych

```python
df2 = df.corr()[['D']]

vals = np.around(df2.values,2)
normal = plt.normalize(-0.5, 0)

fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111, frameon=True, xticks=[], yticks=[])

the_table=plt.table(cellText=vals, rowLabels=df2.index, colLabels=df2.columns, 
                    colWidths = [0.06]*vals.shape[1], loc='center', 
                    cellColours=plt.cm.hot(normal(vals)))
the_table.set_fontsize(14)
the_table.scale(3, 3)
```

Na podstawie powyższej tabelki można stwierdzić, iż istnieje statystycznie istotna korelacja między wartościami w kolumnach C* a wartością klasy decyzyjnej

Następnie sprawdzam korelacje między kolumnami C*

```python
df2 = df.drop('D', 1).corr()

vals = np.around(df2.values,2)
normal = plt.normalize(-0.1, 0.1)

fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111, frameon=True, xticks=[], yticks=[])

the_table=plt.table(cellText=vals, rowLabels=df2.index, colLabels=df2.columns, 
                    colWidths = [0.06]*vals.shape[1], loc='center', 
                    cellColours=plt.cm.hot(normal(vals)))
the_table.set_fontsize(12)
the_table.scale(1.5, 2)
```

Tym razem możemy stwierdzić brak statystycznie istotnej korelacji między kolumnami. Wobec tego, użycie klasyfikatora działającego liniowo względem danych powinno być wystarczające.

Poniżej przedstawiam przebieg i wyniki badań. GridSearchCV jest funkcją, która automatycznie dopasowuje najlepsze parametry z podanych list na podstawie walidacji krzyżowej.

Jakość ostatecznego wyniku dla każdego z algorytmów zbadałem na zbiorze testowym.

```python
results = pd.DataFrame()
```

```python
knc = KNeighborsClassifier()
parameters = {'weights':('uniform', 'distance'), 'algorithm':('auto', 'ball_tree', 'kd_tree', 'brute')}

clf = grid_search.GridSearchCV(knc, parameters, n_jobs=-1)
clf.fit(train_X, train_Y)
score = clf.score(test_X, test_Y)
results = results.append({'Klasyfikator':'kNN', 'Wynik':score, 'Parametry':str(clf.best_params_)}, ignore_index=True)
```

```python
gnb = GaussianNB()
gnb.fit(train_X, train_Y)
score = gnb.score(test_X, test_Y)
results = results.append({'Klasyfikator':'Naive Bayes', 
                          'Wynik':score}, ignore_index=True)
```

```python
dtc = tree.DecisionTreeClassifier()

parameters = {
    'max_depth': (50, 100, 150),
    'min_samples_split': (1, 2, 3),
    'min_samples_leaf': (1, 2, 3)
}

clf = grid_search.GridSearchCV(dtc, parameters, n_jobs=-1)
clf.fit(train_X, train_Y)
score = clf.score(test_X, test_Y)
results = results.append({'Klasyfikator':'Drzewo decyzyjne', 
                          'Wynik':score, 'Parametry':str(clf.best_params_)}, ignore_index=True)
```

```python
lr = linear_model.LogisticRegression()
parameters = {'penalty':('l1', 'l2'), 'C':[0.00001, 0.00005, 0.0001, 0.0005, 0.001]}

clf = grid_search.GridSearchCV(lr, parameters, n_jobs=-1)
clf.fit(train_X, train_Y)
score = clf.score(test_X, test_Y)
results = results.append({'Klasyfikator':'Regresja logistyczna', 
                          'Wynik':score, 'Parametry':str(clf.best_params_)}, ignore_index=True)
```

```python
mlp = MLPClassifier(max_iter=1000, hidden_layer_sizes=(10))

parameters = {
    'algorithm':('l-bfgs', 'sgd'),
    'alpha':(1e-2, 1e-3, 1e-4),
    'learning_rate':('invscaling', 'adaptive'),

}
clf = grid_search.GridSearchCV(mlp, parameters, n_jobs=-1)
clf.fit(train_X, train_Y)
score=clf.score(test_X, test_Y)

results = results.append({'Klasyfikator':'Sieć neuronowa (10)', 
                          'Wynik':score, 'Parametry':str(clf.best_params_)}, ignore_index=True)
```
